```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings(action = "ignore")
```

```python
data = pd.read_csv(r"F:\anaconda\data sets\breast-cancer-data.csv")
```

```python
data.head()
```

```python
data.drop("id", axis = 1, inplace = True)
```

```python
data.diagnosis.unique()
```

```python
data.diagnosis = data.diagnosis.replace("M", 1)
data.diagnosis = data.diagnosis.replace("B", 0)
```

```python
data.head()
```

```python
X=data.drop('diagnosis',axis=1)
y=data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=30)

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train,y_train)

print("What is the Testing Accuracy")
print(lr.score(X_test,y_test))
print("What is the Training Accuracy")
print(lr.score(X_train,y_train))
```

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
scaledxtrain = sc.fit_transform(X_train)
scaledxtest = sc.transform(X_test)

from sklearn.decomposition import PCA

pca = PCA()
xtrain_pca = pca.fit_transform(scaledxtrain)
xtest_pca = pca.transform(scaledxtest)

lr1 = LogisticRegression()
lr1.fit(xtrain_pca, y_train)

print("What is the Testing Accuracy")
print(lr1.score(xtest_pca,y_test))
print("What is the Training Accuracy")
print(lr1.score(xtrain_pca,y_train))
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

### With 2 PCA

```python
pca = PCA(n_components=2)

pca_xtrain = pca.fit_transform(scaledxtrain)
pca_xtest = pca.transform(scaledxtest)

lr2 = LogisticRegression()
lr2.fit(pca_xtrain, y_train)

print("What is the Testing Accuracy")
print(lr2.score(pca_xtest,y_test))
print("What is the Training Accuracy")
print(lr2.score(pca_xtrain,y_train))
```

### With 3 PCA

```python
pca = PCA(n_components=3)

pca_xtrain = pca.fit_transform(scaledxtrain)
pca_xtest = pca.transform(scaledxtest)

lr3 = LogisticRegression()
lr3.fit(pca_xtrain, y_train)

print("What is the Testing Accuracy")
print(lr3.score(pca_xtest,y_test))
print("What is the Training Accuracy")
print(lr3.score(pca_xtrain,y_train))
```
